```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

price_names_data = pd.read_csv("item_names.csv")

price_names_data.head()

# Convert price type to numeric
comma_removed = price_names_data["price"].str.replace(",", "")
numeric_prices = pd.to_numeric(comma_removed)

# preprocessed data. 
df = pd.DataFrame(
    {
        "price": numeric_prices, 
        "item_name": price_names_data["item_name"]
    }
)

sns.set(style="whitegrid")
sns.barplot(x="item_name", y="price", data=df.head())
```
